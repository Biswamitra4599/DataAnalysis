```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Synthetic dataset
np.random.seed(42)
group_A = np.random.normal(loc=50, scale=10, size=200)
group_B = np.random.normal(loc=60, scale=12, size=200)
group_C = np.random.normal(loc=55, scale=9, size=200)

df = pd.DataFrame({
    "Values": np.concatenate([group_A, group_B, group_C]),
    "Group": ["A"]*200 + ["B"]*200 + ["C"]*200
})

###############################################################
# 1 Basic Violin Plot
###############################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values")
plt.title("Basic Violin Plot")
plt.show()

###############################################################
# 2 Violin Plot with Inner Boxplot
###############################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values", inner="box")
plt.title("Violin Plot with Inner Boxplot")
plt.show()

###############################################################
# 3 Violin Plot with Inner Quartiles
###############################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values", inner="quartile")
plt.title("Violin Plot with Quartiles")
plt.show()

###############################################################
# 4 Violin Plot with No Inner Lines (clean look)
###############################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values", inner=None)
plt.title("Clean Violin Plot (No Inner Lines)")
plt.show()

###############################################################
# 5 Split Violin Plot (Two groups compared)
###############################################################

# Create binary groups for split plot
df_split = pd.DataFrame({
    "Score": np.concatenate([group_A, group_B]),
    "Condition": ["A"]*200 + ["B"]*200,
    "Category": ["Group 1"]*200 + ["Group 1"]*200
})

plt.figure(figsize=(7, 5))
sns.violinplot(
    data=df_split,
    x="Category",
    y="Score",
    hue="Condition",
    split=True
)
plt.title("Split Violin Plot (Group A vs Group B)")
plt.show()

###############################################################
# 6 Horizontal Violin Plot
###############################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, y="Group", x="Values", inner="quartile", orient="h")
plt.title("Horizontal Violin Plot")
plt.show()

###############################################################
# 7 Violin + Box Plot Combined
###############################################################

plt.figure(figsize=(8, 5))

sns.violinplot(data=df, x="Group", y="Values", color="lightblue", inner=None)
sns.boxplot(
    data=df, x="Group", y="Values",
    width=0.2, boxprops={'zorder': 3}
)

plt.title("Violin + Box Plot Combined")
plt.show()

###############################################################
# 8 Violin Plot with Custom Bandwidth (KDE smoothing)
###############################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values", bw=0.3)  # smoother KDE
plt.title("Violin Plot with Custom Bandwidth (bw=0.3)")
plt.show()

###############################################################
# 9 Violin Plot using Real Dataset (tips)
###############################################################

tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 5))
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Real Dataset: Violin Plot with Tips Dataset")
plt.show()
```
