```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Synthetic dataset
np.random.seed(42)
group_A = np.random.normal(loc=50, scale=10, size=150)
group_B = np.random.normal(loc=60, scale=8, size=150)
group_C = np.random.normal(loc=55, scale=7, size=150)

df = pd.DataFrame({
    "Values": np.concatenate([group_A, group_B, group_C]),
    "Group": ["A"]*150 + ["B"]*150 + ["C"]*150
})

################################################################
# 1 Basic Strip Plot
################################################################

plt.figure(figsize=(8, 5))
sns.stripplot(data=df, x="Group", y="Values")
plt.title("Basic Strip Plot")
plt.show()

################################################################
# 2 Strip Plot with Jitter (Most Common Variant)
################################################################

plt.figure(figsize=(8, 5))
sns.stripplot(data=df, x="Group", y="Values", jitter=True)
plt.title("Strip Plot with Jitter")
plt.show()

################################################################
# 3 Strip Plot with Custom Jitter Strength
################################################################

plt.figure(figsize=(8, 5))
sns.stripplot(data=df, x="Group", y="Values", jitter=0.3)  # less jitter
plt.title("Strip Plot with Custom Jitter (0.3)")
plt.show()

################################################################
# 4 Horizontal Strip Plot
################################################################

plt.figure(figsize=(8, 5))
sns.stripplot(data=df, y="Group", x="Values", jitter=True, orient="h")
plt.title("Horizontal Strip Plot")
plt.show()

################################################################
# 5 Adding Hue (Color-coded subgroups)
################################################################

# Create subcategories (Male/Female)
subgroup = np.random.choice(["M", "F"], size=450)
df["Subgroup"] = subgroup

plt.figure(figsize=(8, 5))
sns.stripplot(data=df, x="Group", y="Values", hue="Subgroup", jitter=True)
plt.title("Strip Plot with Hue (Two Subgroups)")
plt.show()

################################################################
# 6 Overplotting Control  Using Alpha (transparency)
################################################################

plt.figure(figsize=(8, 5))
sns.stripplot(
    data=df,
    x="Group",
    y="Values",
    jitter=True,
    alpha=0.6           # reduce overlap
)
plt.title("Strip Plot with Transparency to Reduce Overplotting")
plt.show()

################################################################
# 7 Combining Strip Plot with Box Plot
################################################################

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Group", y="Values", width=0.3)
sns.stripplot(data=df, x="Group", y="Values", jitter=True, alpha=0.6)
plt.title("Strip Plot + Box Plot")
plt.show()

################################################################
# 8 Combining Strip Plot with Violin Plot
################################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values", inner="quartile")
sns.stripplot(
    data=df, x="Group", y="Values",
    jitter=True, color='black', size=2
)
plt.title("Strip Plot Over Violin Plot")
plt.show()

################################################################
# 9 Real Dataset Example  Tips Dataset
################################################################

tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 5))
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", jitter=True)
plt.title("Strip Plot  Real Dataset (Tips)")
plt.show()

################################################################
#  Advanced: Split Strip Plot by Hue (dodge)
################################################################

plt.figure(figsize=(8, 5))
sns.stripplot(
    data=df,
    x="Group",
    y="Values",
    hue="Subgroup",
    dodge=True,       # separate dots by hue
    jitter=True,
    size=4
)
plt.title("Strip Plot with Dodge (Separated by Hue)")
plt.show()
```
