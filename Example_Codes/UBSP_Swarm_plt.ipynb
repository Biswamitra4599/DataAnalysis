```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Synthetic dataset
np.random.seed(42)
group_A = np.random.normal(loc=50, scale=10, size=150)
group_B = np.random.normal(loc=60, scale=8, size=150)
group_C = np.random.normal(loc=55, scale=7, size=150)

df = pd.DataFrame({
    "Values": np.concatenate([group_A, group_B, group_C]),
    "Group": ["A"]*150 + ["B"]*150 + ["C"]*150
})

################################################################
# 1 Basic Swarm Plot
################################################################

plt.figure(figsize=(8, 5))
sns.swarmplot(data=df, x="Group", y="Values")
plt.title("Basic Swarm Plot")
plt.show()

################################################################
# 2 Adjusting Marker Size
################################################################

plt.figure(figsize=(8, 5))
sns.swarmplot(data=df, x="Group", y="Values", size=4)
plt.title("Swarm Plot with Smaller Dots (size = 4)")
plt.show()

################################################################
# 3 Swarm Plot with Hue (2 subgroups)
################################################################

# Create subcategories (Male/Female)
df["Subgroup"] = np.random.choice(["M", "F"], size=len(df))

plt.figure(figsize=(8, 5))
sns.swarmplot(data=df, x="Group", y="Values", hue="Subgroup")
plt.title("Swarm Plot with Hue (Two Subgroups)")
plt.show()

################################################################
# 4 Dodge Swarm Plot (separate by hue)
################################################################

plt.figure(figsize=(8, 5))
sns.swarmplot(
    data=df,
    x="Group",
    y="Values",
    hue="Subgroup",
    dodge=True,       # separate swarms for hue
    size=5
)
plt.title("Swarm Plot with Dodge (Separated by Hue)")
plt.show()

################################################################
# 5 Horizontal Swarm Plot
################################################################

plt.figure(figsize=(8, 5))
sns.swarmplot(data=df, y="Group", x="Values", orient="h")
plt.title("Horizontal Swarm Plot")
plt.show()

################################################################
# 6 Swarm + Box Plot Combination
################################################################

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="Group", y="Values", width=0.3)
sns.swarmplot(data=df, x="Group", y="Values", color="black", size=3)
plt.title("Swarm Plot Over Box Plot")
plt.show()

################################################################
# 7 Swarm + Violin Plot Combination
################################################################

plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x="Group", y="Values", inner="quartile")
sns.swarmplot(data=df, x="Group", y="Values", color="black", size=3)
plt.title("Swarm Plot Over Violin Plot")
plt.show()

################################################################
# 8 Real Dataset Example (Tips)
################################################################

tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 5))
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Swarm Plot  Real Dataset (Tips)")
plt.show()

################################################################
# 9 Advanced: Adjust Edgecolor + Alpha
################################################################

plt.figure(figsize=(8, 5))
sns.swarmplot(
    data=df,
    x="Group",
    y="Values",
    size=6,
    alpha=0.7,
    edgecolor="gray",
    linewidth=0.5
)
plt.title("Swarm Plot with Custom Edge Colors & Transparency")
plt.show()

################################################################
#  Advanced: Large Dataset Handling (swarm may be heavy)
################################################################
# For very large datasets, you can subsample to avoid overcrowding:

df_large = df.sample(200)  # subsample

plt.figure(figsize=(8, 5))
sns.swarmplot(data=df_large, x="Group", y="Values")
plt.title("Swarm Plot on Large Dataset (Subsampled to Avoid Lag)")
plt.show()
```
