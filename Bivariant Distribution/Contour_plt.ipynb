# Contour Plot (Matplotlib)

## Basic Contour Plot (Matplotlib)

```python
import numpy as np
import matplotlib.pyplot as plt

# Create grid
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x, y)

# Sample function
Z = np.sin(X**2 + Y**2)

plt.figure(figsize=(4, 3))
plt.contour(X, Y, Z, cmap="viridis")
plt.title("Basic Contour Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.colorbar()
plt.show()
```

## Filled Contour Plot (contourf)

```python
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap="plasma")
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

## Contour Plot with Labels on Lines

```python
plt.figure(figsize=(8, 6))
contours = plt.contour(X, Y, Z, cmap="cool")
plt.clabel(contours, inline=True, fontsize=10)
plt.title("Contour Plot with Labels")
plt.colorbar()
plt.show()
```

## Contour Plot With Custom Levels

```python
levels = [-0.9, -0.5, 0, 0.5, 0.9]

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=levels, cmap="viridis")
plt.title("Contour Plot with Custom Levels")
plt.colorbar()
plt.show()
```

## Filled + Line Contours Together

```python
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=20, cmap="coolwarm")
line_contours = plt.contour(X, Y, Z, levels=20, colors="black", linewidths=0.5)
plt.clabel(line_contours, inline=True, fontsize=8)
plt.title("Mixed: Filled + Line Contours")
plt.colorbar()
plt.show()
```

## Heatmap Style Contour (Smooth Gradient)

```python
plt.figure(figsize=(8,6))
plt.contourf(X, Y, Z, levels=200, cmap="inferno")
plt.title("High-Resolution Heatmap Contour")
plt.colorbar()
plt.show()
```

## Seaborn Contour Plot (kdeplot)

```python
import seaborn as sns

data = np.random.multivariate_normal([0,0], [[1,0.4],[0.4,1]], size=3000)

plt.figure(figsize=(8,6))
sns.kdeplot(
    x=data[:,0],
    y=data[:,1],
    levels=20,
    cmap="viridis"
)
plt.title("Seaborn Contour Plot")
plt.show()
```

## Seaborn Filled Contour Plot

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    x=data[:,0],
    y=data[:,1],
    fill=True,
    cmap="coolwarm",
    thresh=0.05
)
plt.title("Seaborn Filled Contour")
plt.show()
```

## 3D Contour Plot (Contour on 3D Surface)

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='viridis')
ax.set_title("3D Contour Plot")
plt.show()
```

## Contour + Surface (Best 3D Visualization)

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap="viridis", alpha=0.7)
ax.contour(X, Y, Z, 20, cmap="magma", offset=-1)

ax.set_zlim(-1, 1)
plt.title("3D Surface + Contour")
plt.show()
```

## Multiple Contours on Same Axes

```python
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, cmap="cool", alpha=0.7)
plt.contour(X, Y, -Z, cmap="plasma", alpha=0.7)
plt.title("Two Contour Sets on Same Plot")
plt.show()
```

## Contour Plot With Highlighted Level (Important)

```python
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=[0], colors="red", linewidths=2)
plt.contour(X, Y, Z, levels=20, cmap="gray")
plt.title("Contour Plot with Highlighted Level (Z=0)")
plt.show()
```
