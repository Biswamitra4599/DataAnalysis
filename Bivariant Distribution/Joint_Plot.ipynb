# Joint Plots in Python (Scatter + Histograms)

## Requirements

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

## We’ll use the Iris dataset for demonstration.

```python
# Load sample dataset
iris = sns.load_dataset("iris")
iris.head()
```

## Variation 1 — Default Joint Plot (Scatter + Histograms)

```python
sns.jointplot(data=iris, x="sepal_length", y="sepal_width")
plt.suptitle("Default Joint Plot", y=1.02)
plt.show()
```

## Variation 2 — With Different Visual Styles (hexbin)

```python
sns.jointplot(data=iris,
              x="petal_length",
              y="petal_width",
              kind="hex",
              color="purple")

plt.suptitle("Hexbin Joint Plot", y=1.02)
plt.show()
```

## Variation 3 — Kernel Density Estimate (KDE)

```python
sns.jointplot(data=iris,
              x="sepal_length",
              y="petal_width",
              kind="kde",
              fill=True,
              cmap="viridis")

plt.suptitle("KDE Joint Plot", y=1.02)
plt.show()
```

## Variation 4 — Add Regression Line

```python
sns.jointplot(data=iris,
              x="petal_length",
              y="petal_width",
              kind="reg",
              scatter_kws={"alpha": 0.7})

plt.suptitle("Regression Joint Plot", y=1.03)
plt.show()
```

## Variation 5 — Different Species with Hue (Advanced)

Seaborn’s jointplot doesn't support hue directly — but we can use JointGrid to achieve this.

```python
g = sns.JointGrid(data=iris,
                  x="sepal_length",
                  y="sepal_width")

sns.scatterplot(data=iris,
                x="sepal_length",
                y="sepal_width",
                hue="species",
                ax=g.ax_joint)

sns.histplot(data=iris,
             x="sepal_length",
             hue="species",
             multiple="stack",
             ax=g.ax_marg_x)

sns.histplot(data=iris,
             y="sepal_width",
             hue="species",
             multiple="stack",
             ax=g.ax_marg_y)

plt.suptitle("Joint Plot with Hue (JointGrid)", y=1.05)
plt.show()
```

## Bonus: Pairwise Joint Plots in a Grid (PairGrid)

```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.suptitle("Pairwise Joint Scatter + Histograms", y=1.02)
plt.show()
```
