# Matplotlib Hexbin (Most Standard)

## Basic Hexbin

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.random.randn(10000)
y = np.random.randn(10000)

plt.hexbin(x, y, gridsize=30)
plt.colorbar(label='count')
plt.show()
```

## Changing Grid Size & Color Map

```python
plt.hexbin(x, y, gridsize=50, cmap='viridis')
plt.colorbar()
plt.show()
```

## Log-scale Hexbin

```python
plt.hexbin(x, y, gridsize=40, bins='log', cmap='plasma')
plt.colorbar(label='log10(N)')
plt.show()
```

## Hexbin with Extents

```python
plt.hexbin(x, y, gridsize=30, extent=[-3, 3, -3, 3], cmap='inferno')
plt.colorbar()
plt.show()
```
