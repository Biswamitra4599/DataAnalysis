## GENERAL EXPONENTIAL FUNCTION ‚Äî 
ùëé
ùë•
a
x

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 500)

# Bases
a1, a2, a3 = 2, 3, 0.5

y1 = a1**x
y2 = a2**x
y3 = a3**x

plt.figure(figsize=(7,5))
plt.plot(x, y1, label="2^x")
plt.plot(x, y2, label="3^x")
plt.plot(x, y3, label="0.5^x")
plt.title("General Exponential Function: a^x")
plt.xlabel("x")
plt.ylabel("a^x")
plt.grid()
plt.legend()
plt.show()
```

## NATURAL EXPONENTIAL ‚Äî 
ùëí^ùë•


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 500)
y = np.exp(x)

plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.title("Natural Exponential Function: e^x")
plt.xlabel("x")
plt.ylabel("e^x")
plt.grid()
plt.show()
```

## LOGARITHMIC FUNCTION ‚Äî 
logùëé(ùë•)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 10, 500)

# Bases
a1, a2, a3 = 2, 3, 10

y1 = np.log(x) / np.log(a1)   # change of base
y2 = np.log(x) / np.log(a2)
y3 = np.log(x) / np.log(a3)

plt.figure(figsize=(7,5))
plt.plot(x, y1, label="log‚ÇÇ(x)")
plt.plot(x, y2, label="log‚ÇÉ(x)")
plt.plot(x, y3, label="log‚ÇÅ‚ÇÄ(x)")
plt.title("Logarithmic Function: log_a(x)")
plt.xlabel("x")
plt.ylabel("log_a(x)")
plt.grid()
plt.legend()
plt.show()

```

## NATURAL LOGARITHM ‚Äî 
ln(ùë•)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 10, 500)
y = np.log(x)

plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.title("Natural Logarithm: ln(x)")
plt.xlabel("x")
plt.ylabel("ln(x)")
plt.grid()
plt.show()
```

## COMBINED EXPONENTIAL & LOGARITHMIC (for comparison)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 5, 500)

y_exp = np.exp(x)
y_ln  = np.log(x)

plt.figure(figsize=(7,5))
plt.plot(x, y_exp, label="e^x")
plt.plot(x, y_ln, label="ln(x)")
plt.title("Comparison: e^x and ln(x)")
plt.xlabel("x")
plt.grid()
plt.legend()
plt.show()
```

## EXPONENTIAL DECAY ‚Äî 
ùëí^‚àíùë•


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 5, 500)
y = np.exp(-x)

plt.figure(figsize=(7,5))
plt.plot(x, y)
plt.title("Exponential Decay: e^(-x)")
plt.grid()
plt.show()
```

## EXPONENTIAL GROWTH vs DECAY

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 5, 500)

y1 = np.exp(x)
y2 = np.exp(-x)

plt.figure(figsize=(7,5))
plt.plot(x, y1, label="Growth: e^x")
plt.plot(x, y2, label="Decay: e^-x")
plt.title("Exponential Growth vs Decay")
plt.grid()
plt.legend()
plt.show()
```

## LOGARITHM REFLECTION ‚Äî Symmetry: 

log(x) vs ùëí^ùë•

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 5, 500)

plt.figure(figsize=(7,5))
plt.plot(x, np.log(x), label="ln(x)")
plt.plot(x, np.exp(x), label="e^x")
plt.title("Logarithmic vs Exponential")
plt.grid()
plt.legend()
plt.show()
```
