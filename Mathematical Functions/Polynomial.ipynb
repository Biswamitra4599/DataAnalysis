## POLYNOMIAL FUNCTIONS
Linear, Quadratic, Cubic, General Polynomial

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# Polynomial Functions
# ---------------------------
x = np.linspace(-10, 10, 500)

# Linear
y1 = 2*x + 1

# Quadratic
y2 = x**2 - 4*x + 3

# Cubic
y3 = x**3 - 3*x**2 + 2*x

# General Polynomial: 2x^4 - x^3 + 0.5x^2 - x + 1
y4 = 2*x**4 - x**3 + 0.5*x**2 - x + 1

# Plot each one
plt.figure(figsize=(6,4))
plt.plot(x, y1)
plt.title("Linear: 2x + 1")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y2)
plt.title("Quadratic: x^2 - 4x + 3")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y3)
plt.title("Cubic: x^3 - 3x^2 + 2x")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y4)
plt.title("General Polynomial: 2x^4 - x^3 + 0.5x^2 - x + 1")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)

# Avoid division by zero
x1 = x.copy()
x1[np.isclose(x1, 1)] = np.nan

x2 = x.copy()
x2[np.isclose(x2, 0)] = np.nan

# Rational function 1: (x^2 - 1)/(x - 1)
y_r1 = (x1**2 - 1)/(x1 - 1)

plt.figure(figsize=(6,4))
plt.plot(x, y_r1)
plt.title("Rational: (x^2 - 1) / (x - 1)")
plt.axvline(1, linestyle='--')  # discontinuity
plt.grid()
plt.show()

# Rational function 2: 1/x
y_r2 = 1/x2

plt.figure(figsize=(6,4))
plt.plot(x, y_r2)
plt.title("Rational: 1/x")
plt.axvline(0, linestyle='--')  # asymptote
plt.grid()
plt.ylim(-10, 10)
plt.show()
```

## RADICAL / ROOT FUNCTIONS
Square root
Cube root
Higher roots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)

# Square root (only real for x >= 0)
y_sqrt = np.where(x >= 0, np.sqrt(x), np.nan)

# Cube root (real for all x)
y_cbrt = np.cbrt(x)

# 4th root (only x >= 0)
y_root4 = np.where(x >= 0, x**0.25, np.nan)

# Plot sqrt
plt.figure(figsize=(6,4))
plt.plot(x, y_sqrt)
plt.title("Square Root: sqrt(x)")
plt.grid()
plt.show()

# Plot cube root
plt.figure(figsize=(6,4))
plt.plot(x, y_cbrt)
plt.title("Cube Root: cbrt(x)")
plt.grid()
plt.show()

# Plot 4th root
plt.figure(figsize=(6,4))
plt.plot(x, y_root4)
plt.title("Fourth Root: x^(1/4)")
plt.grid()
plt.show()
```

## HIGH-DEGREE POLYNOMIALS (Degree 5, 7, 10)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 800)

# Degree 5 polynomial
y5 = x**5 - 3*x**3 + x

plt.figure(figsize=(6,4))
plt.plot(x, y5)
plt.title("Degree 5 Polynomial: x^5 - 3x^3 + x")
plt.grid()
plt.show()

# Degree 7 polynomial
y7 = x**7 - x

plt.figure(figsize=(6,4))
plt.plot(x, y7)
plt.title("Degree 7 Polynomial: x^7 - x")
plt.grid()
plt.show()

# Degree 10 polynomial
y10 = x**10 - 5*x**6 + 3*x**2 - 1

plt.figure(figsize=(6,4))
plt.plot(x, y10)
plt.title("Degree 10 Polynomial: x^10 - 5x^6 + 3x^2 - 1")
plt.grid()
plt.show()
```

## RANDOM POLYNOMIAL GENERATOR (different result every run)

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # remove if you want random every time

degree = 8
coeffs = np.random.randint(-3, 4, degree + 1)

x = np.linspace(-2, 2, 600)
y = np.polyval(coeffs, x)

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title(f"Random Polynomial (degree {degree})\nCoefficients = {coeffs}")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-6, 6, 2000)
mask = np.isclose(x, 1, atol=1e-3) | np.isclose(x, -2, atol=1e-3)
x_safe = x.copy()
x_safe[mask] = np.nan

y = x_safe / ((x_safe - 1) * (x_safe + 2))

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Rational Function: x / ((x-1)(x+2))")
plt.axvline(1, linestyle='--')
plt.axvline(-2, linestyle='--')
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 7, 2000)
x_safe = x.copy()

# Holes and asymptotes
x_safe[np.isclose(x_safe, 2, atol=1e-3)] = np.nan
x_safe[np.isclose(x_safe, 3, atol=1e-3)] = np.nan

y = ((x_safe-2)*(x_safe+1)) / ((x_safe-2)*(x_safe-3))

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Rational Function with Hole at 2 and Asymptote at 3")
plt.axvline(2, linestyle='--')
plt.axvline(3, linestyle='--')
plt.grid()
plt.ylim(-10, 10)
plt.show()
```

## ROOT FUNCTIONS ‚Äî FRACTIONAL POWERS

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 500)

# x^(1/2) ‚Üí square root
y1 = np.where(x>=0, x**0.5, np.nan)

# x^(1/3) ‚Üí cube root
y2 = np.cbrt(x)

# x^(2/3) ‚Üí interesting shape
y3 = np.sign(x) * (np.abs(x)**(2/3))

# x^(1/4)
y4 = np.where(x>=0, x**0.25, np.nan)

plt.figure(figsize=(6,4))
plt.plot(x, y1)
plt.title("Square Root: x^(1/2)")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y2)
plt.title("Cube Root: x^(1/3)")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y3)
plt.title("Fractional Power: x^(2/3)")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y4)
plt.title("Fourth Root: x^(1/4)")
plt.grid()
plt.show()
```

## NEGATIVE POWERS (x‚Åª¬π, x‚Åª¬≤, x‚Åª¬≥)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
x_safe = x.copy()
x_safe[np.isclose(x_safe, 0)] = np.nan

y1 = 1/x_safe
y2 = 1/(x_safe**2)
y3 = 1/(x_safe**3)

plt.figure(figsize=(6,4))
plt.plot(x, y1)
plt.title("Negative Power: x^(-1)")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y2)
plt.title("Negative Power: x^(-2)")
plt.grid()
plt.show()

plt.figure(figsize=(6,4))
plt.plot(x, y3)
plt.title("Negative Power: x^(-3)")
plt.grid()
plt.show()

```
