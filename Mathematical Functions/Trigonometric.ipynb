## Helper to prevent infinite spikes

```python
import numpy as np
import matplotlib.pyplot as plt

def safe_div(a, b, threshold=1e6):
    with np.errstate(all='ignore'):
        r = a / b
    r = np.where(np.abs(r) > threshold, np.nan, r)
    return r
```

```python
# Main domain
x = np.linspace(-2*np.pi, 2*np.pi, 2000)
```

## sin(x)

```python
plt.figure(figsize=(8,4))
plt.plot(x, np.sin(x))
plt.title("sin(x)")
plt.axhline(0, linewidth=0.7)
plt.grid(True)
plt.show()
```

## cos(x)

```python
plt.figure(figsize=(8,4))
plt.plot(x, np.cos(x))
plt.title("cos(x)")
plt.axhline(0, linewidth=0.7)
plt.grid(True)
plt.show()
```

## tan(x) with asymptotes

```python
y_tan = safe_div(np.sin(x), np.cos(x))

plt.figure(figsize=(8,4))
plt.plot(x, y_tan)
plt.title("tan(x) (with vertical asymptotes)")
plt.grid(True)

# add asymptote lines at pi/2 + k*pi
for k in range(-4,5):
    plt.axvline(np.pi/2 + k*np.pi, linestyle='--', linewidth=0.6)

plt.ylim(-10, 10)
plt.show()
```

## sec(x) = 1 / cos(x)


```python
sec_x = safe_div(1, np.cos(x))

plt.figure(figsize=(8,4))
plt.plot(x, sec_x)
plt.title("sec(x) = 1/cos(x)")
plt.grid(True)
for k in range(-4,5):
    plt.axvline(np.pi/2 + k*np.pi, linestyle='--', linewidth=0.6)
plt.ylim(-10, 10)
plt.show()
```

## csc(x) = 1 / sin(x)

```python
csc_x = safe_div(1, np.sin(x))

plt.figure(figsize=(8,4))
plt.plot(x, csc_x)
plt.title("csc(x) = 1/sin(x)")
plt.grid(True)
for k in range(-4,5):
    plt.axvline(k*np.pi, linestyle='--', linewidth=0.6)
plt.ylim(-10, 10)
plt.show()
```

## cot(x) = cos(x)/sin(x)

#

```python
cot_x = safe_div(np.cos(x), np.sin(x))

plt.figure(figsize=(8,4))
plt.plot(x, cot_x)
plt.title("cot(x) = cos(x)/sin(x)")
plt.grid(True)
for k in range(-4,5):
    plt.axvline(k*np.pi, linestyle='--', linewidth=0.6)
plt.ylim(-10, 10)
plt.show()
```

## Inverse trig functions

```python
x_inv = np.linspace(-1, 1, 500)
```

## arcsin

```python
plt.figure(figsize=(8,4))
plt.plot(x_inv, np.arcsin(x_inv))
plt.title("arcsin(x)")
plt.grid(True)
plt.show()
```

## arccos

```python
plt.figure(figsize=(8,4))
plt.plot(x_inv, np.arccos(x_inv))
plt.title("arccos(x)")
plt.grid(True)
plt.show()
```

## arctan (domain = all real)

```python
x_wide = np.linspace(-10, 10, 500)

plt.figure(figsize=(8,4))
plt.plot(x_wide, np.arctan(x_wide))
plt.title("arctan(x)")
plt.grid(True)
plt.show()
```
