## Absolute Value Function: 

∣x∣

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
y_abs = np.abs(x)

plt.figure(figsize=(7,4))
plt.plot(x, y_abs)
plt.title("Absolute Value Function  |x|")
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid()
plt.show()
```

## Heaviside Step Function: H(x)

Definition (NumPy default):

H(x)=	​

0,
0.5,
1,
	​

x<0
x=0
x>0
	​


```python
y_heaviside = np.heaviside(x, 0.5)

plt.figure(figsize=(7,4))
plt.plot(x, y_heaviside)
plt.title("Heaviside Step Function  H(x)")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

Standard Step Function

step(x)={
0
1
	​

x<0
x≥0
	​


```python
step = np.where(x >= 0, 1, 0)

plt.figure(figsize=(7,4))
plt.plot(x, step)
plt.title("Step Function")
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## Piecewise-defined Function (NumPy)

Example:

f(x)=
	​

x
2
,
x,
3,
	​

x<0
0≤x<2
x≥2
	​


```python
y_piecewise = np.piecewise(
    x,
    [x < 0, (x >= 0) & (x < 2), x >= 2],
    [lambda x: x**2,
     lambda x: x,
     lambda x: 3]
)

plt.figure(figsize=(7,4))
plt.plot(x, y_piecewise)
plt.title("Piecewise Function (NumPy)")
plt.grid()
plt.show()
```

## Smooth Approximation of Heaviside (Advanced)

Useful for ML, PDEs, optimization.


Hϵ(x)=1/(1+e^(−x/ϵ))
	​


```python
epsilon = 0.3
smooth_H = 1 / (1 + np.exp(-x/epsilon))

plt.figure(figsize=(7,4))
plt.plot(x, np.heaviside(x, 0.5), label="Heaviside")
plt.plot(x, smooth_H, label="Smooth Approximation")
plt.title("Smooth Heaviside Approximation")
plt.legend()
plt.grid()
plt.show()
```

## SymPy Piecewise (Symbolic + Numeric)

Symbolic definition:

f(x)=

−x; x<−2

x^2; −2≤x≤2

4; x>2
	​


```python
import sympy as sp

x_sym = sp.symbols('x')

f = sp.Piecewise(
    (-x_sym, x_sym < -2),
    (x_sym**2, (x_sym >= -2) & (x_sym <= 2)),
    (4, True)
)

sp.plot(f, (x_sym, -5, 5), title="Piecewise Function (SymPy)")
```

## Discontinuity Visualization (Jump Function)

f(x)={
	
1; x<0

3; x≥0
	​


```python
y_jump = np.where(x < 0, 1, 3)

plt.figure(figsize=(7,4))
plt.plot(x, y_jump)
plt.scatter([0], [1], color='red')
plt.scatter([0], [3], color='red')
plt.title("Jump Discontinuity")
plt.grid()
plt.show()
```

## Absolute Value: Alternative Definition using Piecewise

∣x∣={
−x, x<0

x, x≥0
	​


```python
y_abs_pw = np.piecewise(x, [x < 0, x >= 0], [lambda x: -x, lambda x: x])

plt.figure(figsize=(7,4))
plt.plot(x, y_abs_pw)
plt.title("Absolute Value using Piecewise")
plt.grid()
plt.show()
```

## Advanced Composite Piecewise Function

f(x)=

sin(x), x<−π

x^2/5, −π≤x<2

e^ -x, x≥2
	​


```python
y_advanced = np.piecewise(
    x,
    [x < -np.pi, (x >= -np.pi) & (x < 2), x >= 2],
    [lambda x: np.sin(x),
     lambda x: x**2 / 5,
     lambda x: np.exp(-x)]
)

plt.figure(figsize=(7,4))
plt.plot(x, y_advanced)
plt.title("Advanced Piecewise Function")
plt.grid()
plt.show()
```
